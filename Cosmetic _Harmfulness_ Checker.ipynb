```python
import pandas as pd
import requests

# Read the CSV file
data = pd.read_csv('cosmetics.csv')

# Function to find ingredients for a given cosmetic
def find_ingredients_for_cosmetic(cosmetic_name):
    cosmetic_row = data[data['Cosmetics'] == cosmetic_name]
    if not cosmetic_row.empty:
        return cosmetic_row['Ingredients'].iloc[0]
    return None

# Get the cosmetic name from the user
user_cosmetic = input("Enter the name of the cosmetic: ")

# Find ingredients for the user-provided cosmetic
ingredients = find_ingredients_for_cosmetic(user_cosmetic)

if ingredients:
    # Connect to the ChatGPT API
    endpoint = 'https://api.openai.com/v1/engines/davinci/completions'
    headers = {
        'Content-Type': 'application/json',
        'Authorization': 'Bearer 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'  
    }

    # Ask if the ingredients are harmful
    prompt = f"I have a cosmetic product called '{user_cosmetic}' with ingredients: {ingredients}. Are these ingredients harmful?"
    data = {
        'prompt': prompt,
        'max_tokens': 100
    }

    # Make the request to ChatGPT API
    response = requests.post(endpoint, headers=headers, json=data)

    if response.status_code == 200:
        answer = response.json()['choices'][0]['text'].strip()
        print(f"Q: {prompt}\nA: {answer}\n")
    else:
        print("Failed to get a response from the ChatGPT API.")
else:
    print("Cosmetic not found in the dataset.")
```
